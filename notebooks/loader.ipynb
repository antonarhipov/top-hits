```kotlin
USE {
    dependencies {
        implementation("org.springframework:spring-jdbc:6.1.4")
        implementation("org.postgresql:postgresql:42.5.4")
    }
}
```

```kotlin
import org.springframework.jdbc.core.JdbcTemplate
import org.springframework.jdbc.datasource.DriverManagerDataSource

val dataSource = DriverManagerDataSource().apply {
    setDriverClassName("org.postgresql.Driver")
    url = "jdbc:postgresql://localhost:5432/mydatabase"
    username = "myuser"
    password = "secret"
}

val jdbcTemplate = JdbcTemplate(dataSource)
```

```kotlin
import org.springframework.jdbc.core.RowMapper
import java.sql.ResultSet

// Create a RowMapper to map database rows to Track objects
val trackRowMapper = RowMapper { rs: ResultSet, _: Int ->
    "${rs.getString("track_name")} - ${rs.getString("artist_name")}"
}

// Select all tracks from the database
val tracks = jdbcTemplate.query("SELECT * FROM tracks ORDER BY id", trackRowMapper)
tracks.take(10).forEach {
    println(it)
}
```
